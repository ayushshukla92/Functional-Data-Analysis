```python
import numpy as np
```

```python
l = []
```

```python
for i in xrange(24):
    l.append( (meanTempDay[15*i:15*i+15].mean() ) )
```

```python
l
```

```python
fifteenDaysMean = pd.DataFrame(l)
```

```python
fifteenDaysMean
```

```python
import matplotlib.pyplot as plt
x = np.linspace(0,23,24)
y = np.array(l)
```

```python
plt.plot(x, y, 'ro', ms= 5)
plt.xlabel('15 days mean')
plt.ylabel('temperature')
plt.title(filename)
```

```python
meanTempDay.mean()
```

```python
from scipy.interpolate import UnivariateSpline
spl = UnivariateSpline(x, y)
xs = np.linspace(0, 23, 1000)
plt.plot(xs, spl(xs), 'g', lw=3)




plt.text(1, meanTempDay.max()-0.5,'mu = '+ str( round(meanTempDay.mean(),2) ), horizontalalignment='left',
     verticalalignment='center')
plt.text(1, meanTempDay.max()-1,'sigma = '+ str( round(meanTempDay.std(),2) ), horizontalalignment='left',
     verticalalignment='center')
plt.show()
```
