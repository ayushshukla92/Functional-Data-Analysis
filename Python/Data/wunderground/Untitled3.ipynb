```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (50.0, 50.0)
```

```python
fig, axes = plt.subplots(10,2)
# fig.tight_layout() # Or equivalently,  "plt.tight_layout()"
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)
axes[0][1].plot(x, y)


plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
from pylab import *
import numpy as np
matplotlib.rcParams['figure.figsize'] = (10, 150)
x = np.linspace(0, 2*np.pi, 400)
y = np.sin(x**2)
sns.set(style="whitegrid")

number_of_subplots=10

for i,v in enumerate(xrange(number_of_subplots)):
    v = v+1
    ax1 = subplot(number_of_subplots,2,v)
    ax1.set_ylim()
    sns.regplot(x,y,ax1)

```
